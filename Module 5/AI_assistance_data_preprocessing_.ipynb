Step 1: Remove Duplicates
Check for duplicate tweets based on Tweet ID and Tweet Text.
Drop duplicates if any.

```python
df = df.drop_duplicates(subset=['Tweet ID'], keep='first')
df = df.drop_duplicates(subset=['Tweet Text'], keep='first')  # Optional, based on use case
```

Since there are no missing values (as per df.info()), this step can be skipped.

**Step 3: Normalize Text Format**
Convert Tweet Text to lowercase for consistency.
Remove extra spaces and strip leading/trailing spaces.

```python
df['Tweet Text'] = df['Tweet Text'].str.lower().str.strip()
```

```python
df['Tweet Text'][1]
```

**Step 4: Remove Special Characters, Links, and Hashtags (Text Cleaning)**
Remove URLs (https://..., www.)
Remove mentions (@username)
Remove hashtags (#topic)
Remove non-alphanumeric characters except spaces

```python
import re

def clean_text(text):
    text = re.sub(r'http\S+|www.\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

df['Tweet Text'] = df['Tweet Text'].apply(clean_text)
```
